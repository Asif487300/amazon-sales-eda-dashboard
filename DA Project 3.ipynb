## Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Pandas is used for data manipulation and Matplotlib is used for visualization.

```python
df = pd.read_csv("amazon.csv")   
```

Dataset successfully loaded into DataFrame.

```python
# Step 1: Removing ₹ and commas
df["discounted_price"] = df["discounted_price"].replace('[₹,]', '', regex=True)

# Step 2: Converting to number
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors="coerce")

# Step 3: Check
print(df["discounted_price"].dtype)
```

```python
# Check single column
df["rating"]

# Multiple columns
df[["product_name","rating","discounted_price"]]

# Save dataset
df.to_csv("Detail.csv", index=False)

# Select specific columns
df[["product_name","discounted_price"]]

# Filter products with price > 500
df[df["discounted_price"] > 500][["product_name","discounted_price"]]
```

#### Insight:

Dataset contains product-related features such as price, rating, and category. Initial exploration helps understand structure and data types.

## Step 4: Data Cleaning

#### 1. Check Format

```python
df.info()
```

#### Insight:

Data types are checked to ensure correct analysis.

#### 2. Check Null Values

```python
df.isnull().sum()
```

#### 3. Impute Null Values

```python
df["rating_count"] = df["rating_count"].fillna(0)
```

```python
df.isnull().sum()
```

#### Mean:

```python
df["rating_count"].dtype
```

```python
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].mean())
```

#### Median:

```python
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].median(),inplace=True)
```

#### Mode:

```python
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].mode(),inplace=True)
```

#### Insight:

Missing values were handled using statistical techniques to maintain data quality.

#### 4. Check Duplicates

```python
df.duplicated()
df.duplicated().sum()
```

#### Remove:

```python
df.drop_duplicates(inplace=True)
```

#### Insight:

Duplicate records were removed to avoid bias in analysis.

#### 5. Drop Rows / Columns

```python
df.drop(0, axis=0)          # row delete
df.drop("img_link", axis=1)  # column delete
```

#### 6. Check Outliers

```python
plt.boxplot(df["discounted_price"])
plt.title("Price Boxplot")
plt.show()
```

#### Insight:

Outliers were detected in price, indicating presence of very high-value products.

## Step 5: EDA (Exploratory Data Analysis)

#### 1. Univariate Analysis

```python
plt.hist(df["discounted_price"])
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

#### Insight:

Most products are in lower price range, with few high-priced items.

#### 2. Bivariate Analysis

```python
plt.scatter(df["discounted_price"], df["rating"])
plt.xlabel("Price")
plt.ylabel("Rating")
plt.title("Price vs Rating")
plt.show()
```

#### Insight:

No strong relationship between price and rating. Expensive products are not always highly rated.

#### 3. Multivariate Analysis

```python
plt.figure(figsize=(8,5))

plt.scatter(df["discounted_price"], df["rating"], alpha=0.5)

plt.xlabel("Price")
plt.ylabel("Rating")
plt.title("Price vs Rating")

plt.grid()

plt.show()
```

#### Insight:

##### A scatter plot was used to analyze the relationship between product price and rating.
##### The visualization shows that there is no strong correlation between price and rating.
##### Both low-priced and high-priced products have a wide range of ratings.

#### Category Analysis

```python
plt.figure(figsize=(10,5))

df["category"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.xticks(rotation=45)

plt.show()
```

#### A bar chart was used to analyze the distribution of products across different categories.
##### The results show that a few categories contain a significantly higher number of products compared to others.
##### This indicates that certain categories dominate the platform, reflecting higher product availability and possibly greater consumer demand.
